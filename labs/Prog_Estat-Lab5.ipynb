```python
import numpy as np
```

```python
def bernoulli(p):
  return 1 if np.random.uniform() <=p else 0

def exponencial(lmda=1):
  u = np.random.uniform()
  return float(-np.log(u)/lmda)
c = np.sqrt((2*np.e)/np.pi)

def h_normal():
  u= np.random.uniform()
  y = exponencial(1)
  while u > (2/np.sqrt(2*np.pi))*np.e**((-y**2)/2)/ c*np.e**y:
    y = exponencial(1)
    u = np.random.uniform()
  return y

def normal_padrao():
  return h_normal() if bernoulli(1/2) else -h_normal()

def normal(mu=0, sig2=1):
  return normal_padrao()*np.sqrt(sig2) + mu
```

```python
# o desafio era manter tudo dentro de um loop, exemplo 12 precisou ser lidado
# à parte e está um show de horrores

def simulacao_MC(func, *densidades, range_=10000):
    return float(sum([func(*densidades) for _ in range(range_)])) / range_

def funcao_12(densidade):
  a = densidade()
  return a,a**2

def funcao_16(densidade):
  x = densidade()
  return np.sin(x)/(x*np.exp(-x))
i=9
for function,*densities in [
    (lambda x: np.exp(-float(x())**2), np.random.uniform),
    (lambda x: 1 if x() > 1 else 0, normal),
    (lambda x,y: 4 if (x()**2+y()**2) < 1 else 0, np.random.uniform, np.random.uniform),
    (0,0),
    (lambda x,y: np.exp(-(x()**2 + y()**2)), np.random.uniform,np.random.uniform),
    (lambda x: np.exp(-x()), exponencial),
    (lambda x: 1/(1+x()), exponencial),
    (funcao_16, exponencial)

  ]:
  if i == 12:
    amostra = [funcao_12(np.random.uniform) for _ in range(10000)]
    media = sum ([i[0] for i in amostra])/10000
    media2 = sum([i[1] for i in amostra])/10000
    var = media2 - media**2
    print(f"Valores estimados para o exemplo 12: média {media} e variância {var}")
  else:
    print(f"valor estimado para o Exemplo {i}: ", simulacao_MC(function,*densities))
  i+=1
```
